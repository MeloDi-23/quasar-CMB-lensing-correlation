```python
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
map_data = np.load('../catalogue/CMB_smoothed_6.npy')   # the 6 arcsec smoothed map
```

```python
l, b, kappa = map_data
```

```python
phi = l
theta = np.pi/2 - b
```

```python
Nside = 2048
```

```python
map = np.ones(hp.nside2npix(Nside)) * hp.UNSEEN
```

```python
index = hp.ang2pix(Nside, theta, phi)
```

```python
map[index] = kappa
```

```python
hp.mollview(map, title='Planck $\\kappa$ map [6 arcmin smoothed]', coord='G')
hp.graticule()
```
