Drawing the corner from numpy.rec file. The benefit is that variable name is in file. Burnin is already applied.

```python
import numpy as np
import corner
```

```python
result = np.load('./np_1.npy')
array_2d = result.view((np.float64, len(result.dtype.names)))
```

```python
array_2d.shape
```

```python
fig = corner.corner(array_2d[:,0:2], labels=result.dtype.names, show_titles=True, color='orange', truths=np.percentile(array_2d, 50, axis=0)[:2])
```
