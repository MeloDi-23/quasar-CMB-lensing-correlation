Drawing the corner from HDFBackend. You can apply different burnin to it, and you can get extra information from it.

```python
from emcee.backends import HDFBackend
import numpy as np
import corner
```

```python
backend = HDFBackend('./backend_1.bin', read_only=True)
```

```python
chi2 = -backend.get_log_prob()
```

```python
import matplotlib.pyplot as plt
```

```python
chi2.min()
```

```python
for i in range(0, 40, 5):
    plt.plot(chi2[:,i])
plt.ylim(bottom=10)
plt.semilogy()
```

```python
chain = backend.get_chain(discard=1000, flat=True)
```

```python
np.percentile(chain, 50, axis=0)
```

```python
chain = backend.get_chain(discard=1000, flat=True)
```

```python
names = ('lgMmin', 'sig_lgM', 'lgM0', 'lgM1p', 'alpha')
```

```python
fig = corner.corner(chain, labels=names, show_titles=True, color='orange', truths=np.percentile(chain, 50, axis=0))
```
