```python
import multiprocessing as mp
from sklearn.neighbors import KDTree
import numpy as np
```

```python
random_points = np.random.uniform((0, 1), size=(1_0_000_000, 2))
query_points = np.random.uniform((0, 1), size=(1_000_000, 2))
```

```python
tree = KDTree(random_points, metric='euclidean')
tree.query(query_points)
```

```python
def func(pos):
    res = tree.query([pos])
    x, y = random_points[res[1]].flatten()
    return x > pos[0] and y > pos[1]

pool = mp.Pool(30)
pool.map(func, query_points)
```

```python
class seal:
    def __init__(self, points):
        self.tree = KDTree(points, metric='euclidean')
        self.points = points
    def func(self, pos):
        res = self.tree.query([pos])
        x, y = self.points[res[1]].flatten()
        return x > pos[0] and y > pos[1]
    def query(self, points):
        pool = mp.Pool(30)
        return np.sum(pool.starmap(self.func, points))
```
