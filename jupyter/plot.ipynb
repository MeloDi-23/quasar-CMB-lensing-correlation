```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy import units as u
```

```python
dat = pd.read_csv('../results/result')
dat_r = pd.read_csv('../results/result_random')

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
# ax.plot(dat['r'], dat['sigma']-dat_r['sigma'], 'o')
ax.errorbar(dat['r'], dat['sigma']*dat['r'], dat['sigma_err']*dat['r'], fmt='o', capsize=5)
ax.errorbar(dat_r['r'], dat_r['sigma']*dat_r['r'], dat_r['sigma_err']*dat_r['r'], fmt='o', capsize=5)
# ax.plot(dat['r'], dat['r']*dat['sigma'])
ax.plot(dat['r'], np.zeros_like(dat['r']), '--')
ax.set_xlabel('$r[\mathrm{Mpc/h}]$')
ax.set_ylabel('$r\Sigma[M_\\odot Mpc/pc^2 h]$')
# ax.loglog()
ax.semilogx()
```

```python
dat = pd.read_csv('../results/result_r=3_100_15_low')
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
# ax.plot(dat['r'], dat['sigma']-dat_r['sigma'], 'o')
ax.errorbar(dat['r'], dat['sigma']*dat['r'], dat['sigma_err']*dat['r'], fmt='o', capsize=5)
ax.errorbar(dat_r['r'], dat_r['sigma']*dat['r'], dat_r['sigma_err']*dat['r'], fmt='o', capsize=5)
ax.plot(dat['r'], np.zeros_like(dat['r']), '--')
ax.semilogx()
```

```python
dat_r = pd.read_csv('../results/result_r=3_100_15_random_low')
sigma = dat['sigma']-dat_r['sigma']
sigma_err = dat['sigma_err']+dat_r['sigma_err']
```

```python
plt.errorbar(dat['r'], sigma*dat['r'], sigma_err*dat['r'], fmt='o', capsize=5)
plt.semilogx()
```
