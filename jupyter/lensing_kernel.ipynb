```python
from astropy.cosmology import Planck18
```

```python
from astropy.cosmology import FlatLambdaCDM
my_cosmos = FlatLambdaCDM(Om0=0.4, H0=65, Tcmb0=2.725)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(0, 5, 50)
```

```python
cosmos = my_cosmos
z = np.linspace(0, 5, 50)
z_s = 1100
chi = cosmos.angular_diameter_distance(z)*(1+z)
chi_s = cosmos.angular_diameter_distance(z_s)*(1+z_s)
W = 3/2*cosmos.Om0*cosmos.H0**2*(1+z)/cosmos.H(z)*chi*(chi_s-chi)/chi_s
```

```python
plt.plot(z, W.value)
plt.xlim(left=0)
plt.ylim(bottom=0)
```
